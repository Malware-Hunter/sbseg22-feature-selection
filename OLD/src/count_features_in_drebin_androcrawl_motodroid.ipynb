# Contagem de Permissões, Intents e API Calls

**IMPORTANTE:**

Antes de comçar, certifique-se de que os seguintes datastes estão localizados no mesmo diretório deste notebook:

- [Androcrawl_LIMPO_apiCalls_intents_permissions.csv](https://github.com/Malware-Hunter/datasets/blob/main/Androcrawl_LIMPO_apiCalls_intents_permissions.csv)
- [Drebin_215_CPI_LIMPO.csv](https://github.com/Malware-Hunter/datasets/blob/main/Drebin_215_CPI_LIMPO.csv)
- [Motodroid Reduzido](https://github.com/Malware-Hunter/feature_selection/blob/main/datasets/motodroid.rar)

Alternativamente, você pode renomear o diretório de onde esses datasets estão sendo lidos na função `pd.read_csv()`

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("./Androcrawl_LIMPO_apiCalls_intents_permissions.csv", index_col=0)
df = df.drop(columns=['class'])
```

```python
df
```

```python
print('# apicalls:\t', df.filter(like='Api Call:').shape[1])
print('# permissions:\t', df.filter(like='Permission:').shape[1])
print('# intents:\t', df.filter(like='Intent Filter').shape[1])
print("Total:\t\t", df.shape[1])
```

```python
df = pd.read_csv("./Drebin_215_CPI_LIMPO.csv", index_col=0)
df = df.drop(columns=['class'])
```

```python
df
```

```python
df.columns
```

```python
print('# apicalls:\t', df.filter(items=[col for col in df.columns if not re.match('^[A-Z_]+$|.*intent\.action.*', col)]).shape[1])
print('# permissions:\t', df.filter(regex='^[A-Z_]+$').shape[1])
print('# intents:\t', df.filter(regex='.*intent\.action.*').shape[1])
print("Total:\t\t", df.shape[1])
```
