```python
import pandas as pd
import numpy as np
```

```python
crawl = pd.read_csv('Androcrawl.csv')
```

```python
list(crawl.columns)
```

```python
col = list(crawl.columns)
```

### Lista de permissões:

```python
permissions = []
for i in col:
    if "Permission" in i:
            permissions.append(i)
```

```python
permissions
```

### Lista de Chamadas de API:

```python
APICalls = []
for i in col:
  if "Api" in i:
    APICalls.append(i)
```

```python
APICalls
```

```python
no_APICalls = []
for i in APICalls:
  if "Notification" in i:
    no_APICalls.append(i)

for i in no_APICalls:
  APICalls.remove(i) 
```

```python
APICalls
```

### Lista de Chamadas de Sistema:

```python
systemCalls = []
for i in col:
  if "System Routine" in i:
    systemCalls.append(i)
```

```python
systemCalls
```

### Lista de Intenções:

```python
intents = []
for i in col:
  if "Intent" in i:
    intents.append(i)
```

```python
intents
```

```python
no_intents = []
for i in intents:
  if "Api" in i:
    no_intents.append(i)

for i in no_intents:
  intents.remove(i) 
```

```python
intents
```

### Lista de todas as colunas de interesse: 

```python
crawlResultList = permissions + APICalls + systemCalls +intents
```

```python
crawlResultList
```

## Dataset Final:

```python
crawl["'Detection Ratio'"].value_counts()
```

```python
crawlResult = crawl[crawlResultList]
```

```python
crawlResult["'Detection Ratio'"] = crawl["'Detection Ratio'"]
```

```python
crawlResult["'Detection Ratio'"] = crawlResult["'Detection Ratio'"].replace("Malicious",int(1))
crawlResult["'Detection Ratio'"] = crawlResult["'Detection Ratio'"].replace("Benign",int(0))
```

```python
crawlResult
```

```python
crawlResult["'Detection Ratio'"].value_counts()
```

### Dataset resultante contém 134 features: permissões, chamadas de sistema, chamadas de API e intenções de 14726 amostras (13279 benignos + 1446 maliciosos)

```python
crawlResult.to_csv("2CPI_Androcrawl.csv")
```
