### Lista de permissões:
As permissões de um app ajudam a apoiar a privacidade do usuário protegendo dados restritos e ações restritas

```python
permissions = []
for i in col:
    if "Permission" in i:
            permissions.append(i)

permissions
```

```python
dangerous = []
c2m = []
harmless = []

for i in permissions:
    if "Dangerous" in i:
      dangerous.append(i)

    if "Harmless" in i:
      harmless.append(i)
    
    if "C2M" in i:
      c2m.append(i)

print(len(dangerous), len(c2m), len(harmless))
```

**Lista de permissões: 61 permissões no total**


*   26 Dangerous Permission
*   3 C2M Permission
*   32 Harmless Permission

### Lista de Chamadas de API:
API Calls: As requisições de API são formas de fazer um software conversar com outro e permitir a troca de informações entre eles.

```python
APICalls = []
for i in col:
  if "Api" in i:
    APICalls.append(i)
    
APICalls
```

```python
no_APICalls = []
for i in APICalls:
  if "Notification" in i:
    no_APICalls.append(i)

for i in no_APICalls:
  APICalls.remove(i)


APICalls
```

```python
len(APICalls)
```

**Lista de API Calls: 27 chamadas no total**

### Lista de Intenções:

```python
intents = []
for i in col:
  if "Intent" in i:
    intents.append(i)

intents
```

```python
no_intents = []
for i in intents:
  if "Api" in i:
    no_intents.append(i)

for i in no_intents:
  intents.remove(i) 

intents
```

```python
c2mIntents = []
suspicious = []

for i in intents:
  if "C2M" in i:
    c2mIntents.append(i)
  
  if "Suspicious" in i:
    suspicious.append(i)

print(len(c2mIntents), len(suspicious))
```

**Lista de Intenções: 27 intenções ao total**
*   3 C2M Intent
*   24 Suspicious Intent Filter

### Lista de todas as colunas de interesse: 

```python
crawlResultList = permissions + APICalls + intents
```

```python
crawlResultList
```

## Dataset Final:

```python
crawl["'Detection Ratio'"].value_counts()
```

```python
crawlResult = crawl[crawlResultList]
```

```python
crawlResult["'Detection Ratio'"] = crawl["'Detection Ratio'"]
```

```python
crawlResult["'Detection Ratio'"] = crawlResult["'Detection Ratio'"].replace("Malicious",int(1))
crawlResult["'Detection Ratio'"] = crawlResult["'Detection Ratio'"].replace("Benign",int(0))
```

```python
crawlResult
```

```python
crawlResult["'Detection Ratio'"].value_counts()
```

```python
crawlResult = crawlResult.mask(crawlResult == False, 0)
crawlResult = crawlResult.mask(crawlResult == 'false', 0)
```

```python
crawlResult
```
